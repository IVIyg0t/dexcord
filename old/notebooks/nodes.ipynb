```python
node_count = 1
cost_per_node = 20
wallet_balance = 0
balance = 0
reward = 1
claim_tax = 0.1
days_since_compounded = 0

def can_compound():
  needed = cost_per_node * (1 + claim_tax)
  return balance >= needed

while node_count <= 100:
  days_since_compounded += 1
  balance += reward * node_count

  if can_compound():
    node_count += 1
    wallet_balance += balance - cost_per_node
    balance = 0

    print(f"Nodes: {node_count} \t wallet: {wallet_balance} \t days: {days_since_compounded}")
    days_since_compounded = 0



```
